```python
import re

def parse_lean_definition(lean_code):
    # Regular expression to capture the name, signature, and content
    pattern = r"\bdef\s+(\w+)\s*(.*?)\s*:=\s*(.*)"
    match = re.search(pattern, lean_code, re.DOTALL)
    
    if match:
        name = match.group(1).strip()
        signature = match.group(2).strip()
        content = match.group(3).strip()
        if signature.endswith('Prop'):
            signature = signature[:-4].strip()
        if signature.endswith(':'):
            signature = signature[:-1].strip()
        return name, signature, content
    else:
        # Regular expression to capture the name, signature, and content
        pattern = r"\bdef\s+(\w+)\s*(.*)"
        match = re.search(pattern, lean_code, re.DOTALL)
        if match:
            name = match.group(1).strip()
            return name, '', ''
        return None, None, None

# # Example usage
# lean_code = "def FnLb (f : ℝ → ℝ) (a : ℝ) : Prop :=\n  ∀ x, a ≤ f x"
# result = parse_lean_definition(lean_code)
# if result:
#     name, signature, content = result
#     print("Name:", name)
#     print("Signature:", signature)
#     print("Content:", content)
# else:
#     print("No match found.")


def get_custom_definitions(traced_file):
    premises = traced_file.get_premise_definitions()
    definitions = {}
    for premise in premises:
        if premise['code'].startswith("def "):
            definition = parse_lean_definition(premise['code'])
            if definition != None:
                definitions[definition[0]] = definition[1]
    return definitions

get_custom_definitions(traced_file)
```

```python
# import importlib
# importlib.reload(utils)

from utils.lean_math_utils import *

def split_tactic(tactic):
    tactic = tactic.replace('·', '').strip()
    if "\n" in tactic:
        parts = tactic.split('\n')
        output = []
        for part in parts:
            output.extend(split_tactic(part))
    elif ";" in tactic:
        parts = tactic.split(';')
        output = []
        for part in parts:
            output.extend(split_tactic(part))
    elif '[' in tactic and tactic[-1] == ']' and ',' in tactic:
        pos = tactic.rfind('[')
        prefix = tactic[0:pos]
        hyps = tactic[pos+1:-1]
        if ',' in hyps:
            output = []
            parts = hyps.split(',')
            for part in parts:
                output.append(prefix + '[' + part.strip() + ']')
        else:
            output = [tactic]
    else:
        output = [tactic]
    return output

assert split_tactic('rw [LinearOrder.min_def a]') == ['rw [LinearOrder.min_def a]']
assert split_tactic('if h : a ≤ b\nthen simp [min_def, if_pos h, le_refl]\nelse simp [min_def, if_neg h]; exact le_of_not_le h') == \
    ['if h : a ≤ b',
     'then simp [min_def]',
     'then simp [if_pos h]',
     'then simp [le_refl]',
     'else simp [min_def]',
     'else simp [if_neg h]',
     'exact le_of_not_le h']
assert split_tactic('simp [max_def, if_neg h, le_refl]') == \
    ['simp [max_def]',
     'simp [if_neg h]',
     'simp [le_refl]']
assert split_tactic('· apply le_trans; apply le_max_left a b; apply le_max_left') == \
    ['apply le_trans',
     'apply le_max_left a b',
     'apply le_max_left']


def get_tactic_templates_of_theorem(thm, custom_defs=None):
    tac_templates = []
    traced_tactics = thm.get_traced_tactics()
    for tracted_tac in traced_tactics:
        sub_tactics = split_tactic(tracted_tac.tactic)
        for sub_tactic in sub_tactics:
            tac_template = parse_lean_state_and_tactic(tracted_tac.state_before, sub_tactic, custom_defs)
            tac_templates.append((sub_tactic, tac_template, tracted_tac.state_before))
    return tac_templates

# Returns list of tuples (tac, tac_template, theorem_full_name, traced_file_path)
def get_tactic_templates_of_traced_file(traced_file):
    tac_templates = []
    custom_defs = get_custom_definitions(traced_file)
    premises = traced_file.get_premise_definitions()
    for premise in premises:
        if premise['code'].startswith('theorem') or premise['code'].startswith('lemma'):
            #if 'theorem' in premise['code'] or 'lemma' in premise['code']:
            thm_name = premise['full_name']
            #print("Theorem", thm_name, '\n')
            thm = traced_file.get_traced_theorem(thm_name)
            tacs = get_tactic_templates_of_theorem(thm, custom_defs)
            for tac, template, state_before in tacs:
                tac_templates.append((tac, template, state_before, premise['code'], thm_name, str(traced_file.lean_file.path)))
    return tac_templates

traced_file = traced_repo.get_traced_file('src/algebra/algebra/basic.lean')
get_tactic_templates_of_traced_file(traced_file)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel

class TripletDataset(Dataset):
    def __init__(self, triplets, tokenizer):
        self.triplets = triplets
        self.tokenizer = tokenizer

    def __len__(self):
        return len(self.triplets)

    def __getitem__(self, index):
        q, s, n = self.triplets[index]
        q_encoded = self.tokenizer(q, padding='max_length', max_length=256, truncation=True, return_tensors='pt')
        s_encoded = self.tokenizer(s, padding='max_length', max_length=64, truncation=True, return_tensors='pt')
        n_encoded = self.tokenizer(n, padding='max_length', max_length=64, truncation=True, return_tensors='pt')
        return q_encoded, s_encoded, n_encoded

class TripletLoss(nn.Module):
    def __init__(self, margin=1.0):
        super(TripletLoss, self).__init__()
        self.margin = margin

    def forward(self, q_embed, s_embed, n_embed):
        dist_pos = torch.norm(q_embed - s_embed, p=2, dim=1)
        dist_neg = torch.norm(q_embed - n_embed, p=2, dim=1)
        loss = torch.mean(torch.relu(dist_pos - dist_neg + self.margin))
        return loss

# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model_state = BertModel.from_pretrained('bert-base-uncased')
model_tag = BertModel.from_pretrained('bert-base-uncased')

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)

# Move the model to the GPU
model_state.to(device)
model_tag.to(device)
```

```python
# Create triplet dataset and dataloader
print("Loading triplets ...")
dataset = TripletDataset(triplets, tokenizer)
dataloader = DataLoader(dataset, batch_size=64)

# Set up optimizer and loss function
optimizer_state = torch.optim.Adam(model_state.parameters(), lr=1e-5)
optimizer_tag   = torch.optim.Adam(model_tag.parameters(), lr=1e-5)
triplet_loss = TripletLoss(margin=1.0)

print("Start training ...")
# Training loop
sum_loss = 0
num_loss = 0
num_iteration = 0
num_epochs = 1
for epoch in range(num_epochs):
    for batch in dataloader:
        q_batch, s_batch, n_batch = batch

        # Move the batch tensors to the GPU
        q_input_ids = q_batch['input_ids'].squeeze(1).to(device)
        q_attention_mask = q_batch['attention_mask'].squeeze(1).to(device)
        s_input_ids = s_batch['input_ids'].squeeze(1).to(device)
        s_attention_mask = s_batch['attention_mask'].squeeze(1).to(device)
        n_input_ids = n_batch['input_ids'].squeeze(1).to(device)
        n_attention_mask = n_batch['attention_mask'].squeeze(1).to(device)

        # Forward pass
        q_outputs = model_state(input_ids=q_input_ids, attention_mask=q_attention_mask)
        s_outputs = model_tag(input_ids=s_input_ids, attention_mask=s_attention_mask)
        n_outputs = model_tag(input_ids=n_input_ids, attention_mask=n_attention_mask)

        q_embeddings = q_outputs.last_hidden_state[:, 0, :]
        s_embeddings = s_outputs.last_hidden_state[:, 0, :]
        n_embeddings = n_outputs.last_hidden_state[:, 0, :]

        # Compute loss
        loss = triplet_loss(q_embeddings, s_embeddings, n_embeddings)
        sum_loss += loss.item()
        num_loss += 1
        num_iteration += 1
        if num_loss % 100 == 1:
            print("Iteration", num_iteration, "Loss:", sum_loss / num_loss)

        # Backward pass and optimization
        optimizer_state.zero_grad()
        optimizer_tag.zero_grad()
        loss.backward()
        optimizer_state.step()
        optimizer_tag.step()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```

```python
import numpy as np
import torch
from transformers import BertTokenizer, BertModel
import faiss

model_tac = model_tag

# Example strings (replace with your own data)
tacs = list(tac_template_freq.keys())

# Convert strings to BERT embeddings
print("Creating embeddings for tactics ...")
embeddings = []
num_processed = 0
with torch.no_grad():
    for tac in tacs:
        encoded_input = tokenizer(tac, padding='max_length', max_length=64, truncation=True, return_tensors='pt').to(device)
        outputs = model_tac(**encoded_input)
        embedding = outputs.last_hidden_state[:, 0, :].detach().cpu().numpy()
        embeddings.append(embedding)
        num_processed += 1
        if num_processed % 1000 == 0:
            print(num_processed, "tactics processed")

# Stack embeddings into a single array
embeddings = np.vstack(embeddings)

# Build Faiss index
print("Building Faiss index ...")
index = faiss.IndexFlatL2(embeddings.shape[1])  # L2 distance metric
index.add(embeddings)
faiss.write_index(index, '/home/mcwave/code/axiomatization/lean3/datasets/rag_tactic_templates/faiss_index_bert_embeds-batch64-60k-loss0066.idx')

fout = open('/home/mcwave/code/axiomatization/lean3/datasets/rag_tactic_templates/tac_template_freq.json', 'w')
json.dump(tac_template_freq, fout)
fout.close()
```

```python
# Example query string
tac = 'rw [← neg_add_cancel_left {variable} {variable}]'

def get_similar_tacs(query_string, faiss_index, all_tacs, num_returned=100, verbose=False):
    # Convert query string to BERT embedding
    with torch.no_grad():
        encoded_input = tokenizer(query_string, padding=True, truncation=True, return_tensors='pt').to(device)
        outputs = model_state(**encoded_input)
        query_embedding = outputs.last_hidden_state[:, 0, :].detach().cpu().numpy()

    # Perform nearest neighbor search
    distances, indices = faiss_index.search(query_embedding, num_returned)

    if verbose:
        # Print the nearest neighbors and their distances
        print(f"Nearest neighbors for: {query_string}")
    output = []
    for i in range(num_returned):
        idx = indices[0][i]
        distance = distances[0][i]
        similarity = 1.0 / (1 + distance*distance/1000)
        output.append((all_tacs[idx],similarity))
        if verbose:
            print(f"{i+1}. {all_tacs[idx]} (similarity: {similarity})")
    return output
            
get_similar_tacs(tac, index, tacs, num_returned=200)
```
